```python
def hor_bar_chart(df, x_value, y_value):
    return px.bar(df, x=x_value, y=y_value)

hor_bar_chart(Sales, ["Store"], ["SalesDollars"])    
```

```python
Sales['SalesDate'] = pd.to_datetime(Sales['SalesDate'])
Purchases['ReceivingDate'] = pd.to_datetime(Purchases['ReceivingDate'])


#Wykres Pareto

def pareto_stores(df, groupby_cokumn, value_column):
    top_sales_store = df.groupby([groupby_cokumn])[value_column].sum().reset_index().sort_values(by=[value_column], ascending=False)
    top_sales_store['cumulative_sales'] = top_sales_store['SalesDollars'].cumsum().round(2)
    top_sales_store['pareto'] = top_sales_store['cumulative_sales'] / top_sales_store['cumulative_sales'].max()


    return 


top_sales_store = top_sales_store[top_sales_store['pareto'] <= 0.8]
top_store_list = top_sales_store['Store'].to_list()


Sales_top80= Sales[Sales['Store'].isin(top_store_list)]


top_sales_brand = Sales_top80.groupby(["Brand"])["SalesDollars"].sum().reset_index().sort_values(by=['SalesDollars'], ascending=False)
top_sales_brand['cumulative_sales'] = top_sales_brand['SalesDollars'].cumsum().round(2)
top_sales_brand['pareto'] = top_sales_brand['cumulative_sales'] / top_sales_brand['cumulative_sales'].max()

top_sales_brand = top_sales_brand[top_sales_brand['pareto'] <= 0.8]
top_brand_list = top_sales_brand['Brand'].to_list()

Sales_top80= Sales_top80[Sales_top80['Brand'].isin(top_brand_list)]

display(Sales_top80)
```
